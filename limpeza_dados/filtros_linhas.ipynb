```python
# Importar bibliotecas
import pandas as pd
# Dados de expectativas do IPCA (Focus/BCB)
dados_focus = pd.read_csv(
  filepath_or_buffer = "https://olinda.bcb.gov.br/olinda/servico/Expectativas/versao/v1/odata/ExpectativasMercadoAnuais?%24filter=%28Indicador%20eq%20%27IPCA%27%20or%20Indicador%20eq%20%27IGP-M%27%29%20and%20Data%20ge%20%272023-01-01%27%20and%20Data%20le%20%272023-02-12%27&%24format=text/csv&%24orderby=Data%20desc",
  decimal = ","
  )
```

```python
#Filter one condition
dados_focus.query('Indicador == "IPCA"')
```

```python
# Filter multi condition
data_max=dados_focus['Data'].max()
dados_focus.query('Indicador == "IPCA" and DataReferencia == 2026 and numeroRespondentes > 100')

# Filtrando por mÃºltiplas condiÃ§Ãµes
dados_focus.query("Indicador != 'IPCA' and DataReferencia in [2023, 2024] and Data >= Data.max() and baseCalculo == 0") 
# OR
# Filtrando por mÃºltiplas condiÃ§Ãµes
dados_focus.query("Indicador != 'IPCA' and DataReferencia in [2023, 2024] and Data >= @data_max and baseCalculo == 0")
```
