```python
# Importar bibliotecas
import pandas as pd
# Dados de expectativas de inflação (Focus/BCB)
dados_focus = pd.read_csv(
  filepath_or_buffer = "https://olinda.bcb.gov.br/olinda/servico/Expectativas/versao/v1/odata/ExpectativasMercadoAnuais?%24filter=%28Indicador%20eq%20%27IPCA%27%20or%20Indicador%20eq%20%27IGP-M%27%29%20and%20Data%20ge%20%272023-01-01%27%20and%20Data%20le%20%272023-02-12%27&%24format=text/csv&%24orderby=Data%20desc",
  decimal = ","
  )
```

```python
dados_focus.groupby(by="DataReferencia").head(1)
```

```python
#Define groups and apply line filter
dados_focus.query('baseCalculo == 0').groupby(by=['Indicador','DataReferencia']).apply(lambda x:x.query('Mediana == Mediana.max()'),include_groups=False)
```

```python
dados_focus.query('baseCalculo == 0').groupby(by=['Indicador','DataReferencia'],group_keys=False).apply(lambda x:x.query('Mediana == Mediana.max()'),include_groups=False)
```
